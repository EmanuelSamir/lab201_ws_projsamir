```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
```

```python
model = Sequential()
model.add(Dense(units=10, activation='linear', input_dim=2))
model.add(Dense(units=10, activation='linear'))
model.add(Dense(units=1, activation='linear'))

model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
q_data = df.j7.values.reshape(20001,1)
T_data = df.e7.values.reshape(20001,1)
qd_data = df.jd7
x_train = np.concatenate([q_data, T_data], axis =1)
y_train = qd_data.values.reshape(20001,1)
print x_train.shape
print y_train.shape

#x_train = df['j7', 'e7']
#y_train = df['jd7
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=50)
```
