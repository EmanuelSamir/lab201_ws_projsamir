```python
#importing all libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, LeakyReLU
```

```python
df = pd.read_csv('training_data_joint1.csv')
df_eval = pd.read_csv('evaluation_data_joint1.csv')
df_val = pd.read_csv('validation_data_joint1.csv')
fig1 = plt.figure()
df.j1[0:500].plot()
df.jd1[0:500].plot(); plt.legend(loc='best')
fig2 = plt.figure(); df.e1[0:500].plot(); plt.legend(loc='best')
```

### Prueba de validaci√≥n para modelo FFNN
#### Solo Neuronas Lineales (model1)
Se utiliza un modelo de 1 input layer de 15 neuronas lineales, 2 hidden layers de 15 neuronas lineales y 1 output layer de 2 neuronas lineales.  

```python
# Generating model
model1 = Sequential()
model1.add(Dense(units=10, activation='linear', input_dim=2))
model1.add(Dense(units=10, activation='linear'))
model1.add(Dense(units=10, activation='linear'))
model1.add(Dense(units=1, activation='linear'))

model1.compile(loss='mean_squared_error',
              optimizer='Adam',
              metrics=['accuracy'])

(number_training_data, _) = df.shape
q_tmp = df.j1.values.reshape(number_training_data,1)
T_tmp = df.e1.values.reshape(number_training_data,1)
qd_tmp = df.jd1
x_train = np.concatenate([q_tmp, T_tmp], axis =1)
y_train = qd_tmp.values.reshape(number_training_data,1)
#print x_train.shape
#print y_train.shape
```

#### Feeding data

```python
model1.fit(x_train, y_train, epochs=10, batch_size=50)
```

```python
(number_evaluation_data, _) = df_eval.shape
q_tmp = df_eval.j1.values.reshape(number_evaluation_data,1)
T_tmp = df_eval.e1.values.reshape(number_evaluation_data,1)
qd_tmp = df_eval.jd1
x_eval = np.concatenate([q_tmp, T_tmp], axis =1)
y_eval = qd_tmp.values.reshape(number_evaluation_data,1)
```

```python
model1.evaluate(x_eval, y_eval, batch_size=50)
```

```python
(number_validation_data, _) = df_val.shape
q_tmp = df_val.j1.values.reshape(number_validation_data,1)
T_tmp = df_val.e1.values.reshape(number_validation_data,1)
qd_tmp = df_val.jd1
x_val = np.concatenate([q_tmp, T_tmp], axis =1)
y_val = qd_tmp.values.reshape(number_validation_data,1)
```

```python
y1_pred = model1.predict(x_val)
```

```python
df_val['y1_pred'] = y1_pred
```

```python
fig5 = plt.figure(); df_val[0:20].plot(x = 'e1', y = 'jd1', marker='.',lw=0.3)
df_val[0:20].plot(x = 'e1', y = 'y1_pred', marker='.',lw=0.3); plt.legend(loc='best')
```

```python
fig6 = plt.figure(); df_val[0:1000].plot(x = 'j1', y = 'jd1', marker='.',lw=0.05)
df_val[0:1000].plot(x = 'j1', y = 'y1_pred', marker='.',lw=0.05); plt.legend(loc='best')
```
